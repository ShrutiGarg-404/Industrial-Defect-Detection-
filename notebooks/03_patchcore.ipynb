```python
explanation = """
HOW PATCHCORE WORKS —in plain English:
=========================================
Step 1: TRAINING (on good images only)
  → Feed all good images through ResNet
  → Extract features from each image patch
  → Store these features in a "memory bank"
  → Memory bank = what NORMAL looks like

Step 2: TESTING (on new images)
  → Feed test image through same ResNet
  → Compare each patch to nearest match in memory bank
  → If very DIFFERENT from anything in memory → ANOMALY
  → Build a heatmap: bright = anomalous region

WHY BETTER THAN RESNET18:
  → ResNet18 needed defective examples → we had none
  → PatchCore only needs good examples → perfect for MVTec!
  → PatchCore shows WHERE the defect is → ResNet18 couldn't
"""
print(explanation)
```

```python
from torchvision.transforms import v2 as T

# Define transforms manually since image_size param no longer exists
augmentations = T.Compose([
    T.Resize((224, 224)),
    T.ToTensor(),
])

datamodule = MVTecAD(
    root=str(DATASET_ROOT),
    category=CATEGORY,
    train_batch_size=32,
    eval_batch_size=32,
    num_workers=0,
    augmentations=augmentations,
)

print(f"Datamodule created for: {CATEGORY}")
print(f"Train: {DATASET_ROOT / CATEGORY / 'train' / 'good'}")
print(f"Test:  {DATASET_ROOT / CATEGORY / 'test'}")
```
